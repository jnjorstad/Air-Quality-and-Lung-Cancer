```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
#import tensorflow as tf

import pandas as pd 
df = pd.read_excel("lung-data.xlsx")
df
```

```python
df.columns
```

```python
df_preprocessed = df.drop(columns=['County','State','Status Variable','LTD','Intercept','Slope','Status','Inter',
                                   'Slp','control','treat','Local_Treat','Lung Cancer'])
df_preprocessed
```

```python
df_preprocessed.dtypes
```

```python
df_preprocessed = df_preprocessed.dropna()
```

```python
df_preprocessed_1 = df_preprocessed[df_preprocessed.RT != '*']
df_preprocessed_1
```

```python
df_preprocessed_1['RT'].value_counts()
```

```python
df_preprocessed_1['RT'] = df_preprocessed_1['RT'].replace({'stable': 0, 'rising': 1, 'falling':2})
```

```python
df_preprocessed_1['RT'] = df_preprocessed_1['RT'].astype(int)
```

```python
# Note: this classification task is not directly related to project 4. I was simply checking how different classifier 
# can predict the air quality class ('stable', 'falling' and 'rising') from RT column based on different input features
# Split our preprocessed data into our features and target arrays
y = df_preprocessed_1['RT'].values
X = df_preprocessed_1.drop('RT', axis=1).values
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20,random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_train)
```

```python
from sklearn import metrics
# Model Accuracy on train dataset, how often is the classifier correct?
print("Accuracy:", metrics.accuracy_score(y_train, y_pred))
```

```python
# Model Accuracy on test dataset
y_pred1 = clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred1))
```

```python
confusion_matrix = metrics.confusion_matrix(y_test, y_pred1)
confusion_matrix
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred1))
```

```python
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
clf_svm = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf_svm.fit(X_train, y_train)
y_pred_svm = clf_svm.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_svm))
```

```python
confusion_matrix_svm = metrics.confusion_matrix(y_test, y_pred_svm)
confusion_matrix_svm
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_svm))
```

```python
# Note: All regression tasks are related to Project 4
df_regr = df.drop(columns=['FIPS_code','County','State','Status Variable','LTD','Intercept','Slope','Status','Inter',
                           'Slp','control','treat','Local_Treat','AAC','EQI','RT'])
df_regr
```

```python
from sklearn.ensemble import RandomForestRegressor
# Split our preprocessed data into our features and target arrays
y = df_regr['Lung Cancer'].values
X = df_regr.drop('Lung Cancer', axis=1).values
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 42)
model_regr = RandomForestRegressor(max_depth=6, random_state=42)
model_regr.fit(X_train, y_train)
```

```python
model_regr.feature_importances_
```

```python
# feature_names = [f"feature {i}" for i in range(X.shape[1])]
```

```python
# import pandas as pd
# import matplotlib.pyplot as plt
# plt.barh(feature_names, model_regr.feature_importances_)
```

```python
y_pred_rf = model_regr.predict(X_test)
```

Quantifying RandomForestRegressor model with metrices below 

mean squared error (MSE) and
R2 Score

Note: There are a variety of ways to quantify the model, but MSE and R2 are quite common

```python
from sklearn.metrics import mean_squared_error, r2_score

# Score the prediction with MSE and R2
mse_rf = mean_squared_error(y_test, y_pred_rf, squared=False)
r2_rf = r2_score(y_test, y_pred_rf)

print(f"mean squared error (MSE): {mse_rf}")
print(f"R-squared (R2 ): {r2_rf}")
```

A good MSE score will be close to zero, while a good R2 score will be close to 1.
The obtained value suggest that RandomForestRegressor works efficiently on the dataset

```python
# Five-fold cross validation
from sklearn.model_selection import cross_val_score
cv_results_regr = cross_val_score(model_regr, X, y, cv=5, scoring='r2')
```

```python
cv_results_regr
```

Five-fold cross validation results suggest RandomForestRegressor works efficiently on the dataset
